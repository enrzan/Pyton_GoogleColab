## Instanciando Matrizes

* Nas linguagens MatLab e R, uma matriz é definida com elementos da posição `(1,1)` até a posição `(L, C)`, onde `L` representa o número de linhas e `C` representa o número de colunas de uma matriz.

## Formas de se percorrer uma matriz

* Analogamente ao vetor, uma matriz, após criada, possui tamanho fixo.
* Geralmente, para cada dimensão da matriz, é recomendado usar uma estrutura de repetição `para`,
  > com o objetivo de tornar o código mais compacto e genérico para matrizes de quaisquer dimensões.
* Além disso, é natural percorrer (ou varrer) a matriz
  * da linha `i=0` até a linha `i<L`, onde `L` representa o número de linhas de uma matriz.
  * e da coluna `j=0` até a coluna `j<C`, onde `C` representa o número de colunas de uma matriz.

* No exemplo a seguir em pseudocódigo, uma matriz `m` é criada com `6` elementos, sendo três linhas e `2` colunas, e os dois laços `para` inicializam todos os seus elementos com o valor 0.


```
Pseudocódigo:
Instanciar uma matriz m com 3 linhas e 2 colunas
inteiros L=3, C=2
Para cada i, de i=0; até i<L; passo i=i+1 faça
    Para cada j, de j=0; até j<C; passo j=j+1 faça
        m[i,j] = 0
```

* Existem várias formas de percorrer (ou varrer) uma matriz usando uma estrutura de repetição.
* Por exemplo, é possível percorrer uma matriz
  * do primeiro elemento `m[0,0]`
  > (convencionando como canto superior esquerdo)
  * até o último elemento `m[L-1,C-1]`
    > (convencionando como canto inferior direito da matriz),
  * linha por linha, ou coluna por coluna.
  
* Esse tipo de varredura é chamada **raster**.

* A varredura inversa é chamada **anti-raster**, do último elemento inferior direito até o primeiro elemento superior esquerdo.

Antes de resolver este primeiro exemplo, seguem algumas considerações sobre como criar uma matriz em Python.

````
# Matriz de inteiros iniciada com zero com 3x2 elementos
m1 = [[0, 0], [0, 0], [0, 0]]

# NÃO criar matrizes dessa forma, pois pode resultar em comportamento inesperado
m2 = 3 * [2 * [0]]

# Exibe as matrizes e seus tipos
print("Matriz m1:", m1, "Tipo:", type(m1))
print("Matriz m2:", m2, "Tipo:", type(m2))
````

````
# Criação de uma lista m2 contendo 3 referências para a mesma lista [0, 0]
# Esse é um padrão comum, mas leva a problemas inesperados em Python
m2 = 3 * [2 * [0]]

# Exibe o tipo da variável m2
print("Tipo de m2:", type(m2))

# Modifica o primeiro elemento da primeira lista interna (m2[0][0])
# No entanto, todas as listas internas são referências para a mesma lista
# Portanto, a modificação afeta todas as "cópias" dessa lista
m2[0][0] = 8

# Exibe a variável m2 após a modificação
print("m2 após modificação:", m2)
````


**ATENÇÃO:** Ao fazer m2 = 3*[2*[0]], não cria uma matriz com 3 linhas e 2 colunas, mas sim uma lista com 3 cópias do mesmo objeto, que é uma lista com 2 zeros. Isso significa que, quando você altera um elemento em uma linha, todas as outras linhas também são afetadas, porque elas compartilham a mesma lista interna.

````
# Uma abordagem alternativa que funciona
m2 = [[0] * 2 for _ in range(3)]

# Modificando o primeiro elemento da primeira linha da matriz m2
m2[0][0] = 8

# Exibindo a matriz após a modificação
print("Matriz m2:", m2)
````

````
L, C = 2, 3

# NÃO criar matrizes dessa forma!!!
m1 = [[0]*C]*L
m1[0][0] = 8

# Forma correta de criar uma matriz L x C de zeros!!!
m2 = [[0]*C for i in range(L)]
m2[0][0] = 8
````

````
import random

def criar_matriz_aleatoria(linhas, colunas):
    """
    Cria uma matriz com números inteiros aleatórios no intervalo de 0 a 9.

    Parâmetros:
    - linhas (int): Número de linhas na matriz.
    - colunas (int): Número de colunas na matriz.

    Retorna:
    - matriz (list): Matriz com números inteiros aleatórios.
    """
    matriz = [[0]*colunas for _ in range(linhas)]

    for i in range(linhas):
        for j in range(colunas):
            elemento = random.randint(0, 9)
            matriz[i][j] = elemento

    return matriz

# Exemplo de uso:
L, C = random.randint(2, 9), random.randint(2, 9)
print(f"L={L}, C={C}")
matriz_aleatoria = criar_matriz_aleatoria(L, C)

# Exibindo a matriz criada
escrevaMatriz(matriz_aleatoria)
````

````
def leiaMatriz():
    """
    Lê uma matriz de inteiros.

    Retorna:
    - matriz (list): Matriz de inteiros.
    """
    matriz = []
    ler_linha = input() # Digite os elementos da matriz (pressione Enter para finalizar"

    while ler_linha:
        linha = [int(i) for i in ler_linha.split(' ') if i]
        matriz.append(linha)
        ler_linha = input()

    return matriz

m1 = leiaMatriz() # tem que terminar com uma linha em branco
L = len(m1)  # Número de linhas
C = len(m1[0])  # Número de colunas (assumindo que todas as linhas têm o mesmo comprimento)
print(f"L={L}, C={C}")
escrevaMatriz(m1)
````

````
def leiaMatriz2(matrizTexto):
    """
    Lê uma matriz de inteiros a partir de uma string.

    Parâmetros:
    - matrizTexto (str): String contendo a representação da matriz.

    Retorna:
    - matriz (list): Matriz de inteiros.
    """
    matriz = []
    ler_linhas = matrizTexto.strip().split('\n')

    for linha in ler_linhas:
        if linha:  # Se a linha tiver algum conteúdo
            matriz.append([int(i) for i in linha.split(' ') if i])

    return matriz

matrizTexto = '''
0 9 3 6 9 8 4 5 4
8 1 2 3 5 2 9 9 6
4 1 1 0 9 9 8 2 7
5 2 8 4 6 6 0 8 0
4 7 6 4 3 9 3 3 5
2 6 0 4 0 7 5 5 2
9 8 4 8 4 7 1 4 3
'''

m1 = leiaMatriz2(matrizTexto)
L = len(m1)
C = len(m1[0])  # Obtendo o número de colunas, assumindo que todas as linhas têm o mesmo comprimento
print(f"L={L}, C={C}")
escrevaMatriz(m1)
````

````
def ler_vetor(texto):
    vet_texto = texto.split(" ")
    v = [0] * len(vet_texto)
    n = 0
    for i in range(len(vet_texto)):
        if vet_texto[i] != "":
            v[n] = int(vet_texto[i])
            n += 1
    vet_saida = [0] * n
    for i in range(len(vet_saida)):
        vet_saida[i] = v[i]
    return vet_saida

def leiaMatriz3(texto):
    """
    Lê uma matriz de inteiros a partir de uma string (sem append)

    Parâmetros:
    - matrizTexto (str): String contendo a representação da matriz.

    Retorna:
    - matriz (list): Matriz de inteiros.
    """
    linhas = texto.splitlines()
    num_linhas = len(linhas)
    matriz = [[] for _ in range(num_linhas)]
    contador_linhas = 0
    for i in range(num_linhas):
      linha = linhas[i]
      #print(f"[{linha}]")
      if linha != "":
        matriz[contador_linhas] = ler_vetor(linha)
        contador_linhas += 1
    matriz_saida = [[] for _ in range(contador_linhas)]
    for i in range(contador_linhas):
      matriz_saida[i] = matriz[i]
    return matriz_saida

matrizTexto = '''
0 9 3 6 9 8 4 5 4
8 1 2 3 5 2 9 9 6
4 1 1 0 9 9 8 2 7
5 2 8 4 6 6 0 8 0
4 7 6 4 3 9 3 3 5
2 6 0 4 0 7 5 5 2
9 8 4 8 4 7 1 4 3
'''

m1 = leiaMatriz3(matrizTexto)
L = len(m1)
C = len(m1[0])  # Obtendo o número de colunas, assumindo que todas as linhas têm o mesmo comprimento
print(f"L={L}, C={C}")
escrevaMatriz(m1)
````

## Escrever e Ler Arquivo

Para criar ou abrir um arquivo em Python, usamos o método `open`. Ele aceita dois argumentos principais:

**Nome do Arquivo:** O primeiro argumento é o nome do arquivo que queremos criar ou abrir. Por exemplo, `"meuarquivo.txt"`.

**Modo de Acesso:** O segundo argumento é uma *string* que especifica o modo de acesso ao arquivo. Os modos mais comuns são:

* `"w"` (*write*): Abre o arquivo para escrita. Se o arquivo já existir, ele será sobrescrito. Se não existir, um novo arquivo será criado.

* `"r"` (*read*): Abre o arquivo para leitura. Se o arquivo não existir, ocorrerá um erro.

````
# Escrever em um arquivo
nome_arquivo = "meuarquivo.txt"
conteudo = "Olá, mundo!\nEsta é mais uma linha de exemplo."

with open(nome_arquivo, "w") as arquivo: # escrever em arquivo
    arquivo.write(conteudo)

# Ler de um arquivo
with open(nome_arquivo, "r") as arquivo: # ler arquivo
    linhas = arquivo.read().split('\n')  # Usar split para dividir as linhas
    print(linhas)
    for i in range(len(linhas)):
        print(linhas[i])
````

Para a leitura e escrita de matrizes, é crucial considerar que a leitura deve seguir os padrões definidos durante a escrita. Por exemplo, é importante garantir a consistência nos separadores de elementos, como `","`, e assegurar que não haja linhas vazias no início ou no final da matriz.

````
import random

# Exemplo de uso:
L, C = random.randint(2, 9), random.randint(2, 9)
print(f"L={L}, C={C}")
m = [[random.randint(0, 9) for _ in range(C)] for _ in range(L)]

def escrevaMatriz(m):
    # Exibindo a matriz formatada
    for i in range(len(m)):
        for j in range(len(m[0])):
            print(f"{m[i][j]:<6}", end="")  # Formatando a saída
        print()

def escrevaMatrizCSV(nome_arquivo, m):
    with open(nome_arquivo, "w") as arquivo: # Abrindo o arquivo para escrita
        texto = ""
        for i in range(len(m)):
            for j in range(len(m[0])):
                # Escrevendo o elemento atual da matriz
                texto+=str(m[i][j])

                # Adicionando vírgula entre elementos, exceto para o último
                if j < len(m[0]) - 1:
                    texto+=","

            # Pular linha, exceto para a última linha da matriz
            if i < len(m) - 1:
                texto+='\n'

        arquivo.write(texto)
        print(texto)
    print(f"Matriz escrita em '{nome_arquivo}' com sucesso!")

# Exibindo a matriz criada
escrevaMatrizCSV('matriz.csv', m)
escrevaMatriz(m)
````

````
def leiaMatrizCSV(nome_arquivo):
    # Abrindo o arquivo para leitura
    with open(nome_arquivo, "r") as arquivo:
        # Lendo todas as linhas do arquivo e salvando em uma lista
        linhas = arquivo.read().split('\n')

        # Obtendo o número de linhas e colunas da matriz
        L = len(linhas)
        C = len(linhas[0].split(","))

        # Inicializando uma matriz com zeros
        m = [[0] * C for _ in range(L)]

        # Preenchendo a matriz com os elementos do arquivo
        for i in range(L):
            linha = linhas[i].split(",")  # Dividindo a linha pelos caracteres de vírgula para obter os elementos individuais
            for j in range(C):
                # Verificando se há elementos suficientes na linha antes de atribuir a matriz
                if len(linha) > j:
                    m[i][j] = int(linha[j])  # Convertendo o elemento para inteiro e atribuindo à matriz
    return m

m2 = leiaMatrizCSV('matriz.csv')
escrevaMatriz(m2)
````

````
def saoMatrizesIguais(matriz1, matriz2):
    # Verifica se as matrizes têm o mesmo número de linhas e colunas
    if len(matriz1) != len(matriz2) or len(matriz1[0]) != len(matriz2[0]):
        return False

    # Comparação elemento por elemento
    for i in range(len(matriz1)):
        for j in range(len(matriz1[0])):
            if matriz1[i][j] != matriz2[i][j]:
                return False

    # Se nenhum elemento for diferente, as matrizes são iguais
    return True

saoMatrizesIguais(m, m2)
````

### Escrever e Ler Imagem PBM

O formato PBM (*Portable Bitmap*) é de arquivo de imagem binária (preto e branco) rasterizado.  Utilizando a biblioteca PIL ([*Python Imaging Library*](https://pypi.org/project/pillow/)), é possível criar uma imagem PBM contendo texto de forma simples e eficaz. Neste exemplo, uma nova imagem é criada e o texto 'U F A B C' é desenhado sobre ela. O resultado é então salvo em um arquivo PBM, permitindo fácil compartilhamento e visualização.

````
# criar uma image no formato PBM contendo um texto
from PIL import Image, ImageDraw, ImageFont

# Cria uma nova imagem em branco
largura, altura = 60, 8
imagem = Image.new("1", (largura, altura), color=0)

# Criando um objeto ImageDraw para desenhar na imagem
draw = ImageDraw.Draw(imagem)

# Definindo o texto a ser desenhado
texto = "U F A B C"

# Carregando uma fonte
fonte = ImageFont.load_default()

# Desenhando o texto na imagem
draw.text((4, -2), texto, font=fonte, fill=1)

# Salva a imagem em um arquivo PBM no formato texto
with open("ufabc.pbm", "w") as f:
    f.write("P1\n{} {}\n".format(imagem.width, imagem.height))
    for y in range(imagem.height):
      for x in range(imagem.width):
        f.write(str(imagem.getpixel((x, y))))
      f.write('\n')
````

````
# Mostra conteúdo do arquivo ufabc.pbm
with open('ufabc.pbm', 'r') as f:
    print(f.read())
````

````
# mostra imagem utilizando matplotlib
from PIL import Image
import matplotlib.pyplot as plt

# # Carregando a imagem
image_path = "ufabc.pbm"
image = Image.open(image_path)

# Exibindo a imagem
plt.imshow(image, cmap='gray')
plt.show()
````

O método `leiaImagemPBM` a seguir lê uma imagem no formato PBM. Ele começa verificando se o arquivo é uma imagem PBM válida. Em seguida, extrai as dimensões da imagem e inicializa uma matriz de pixels. Depois, ele percorre as linhas do arquivo, preenchendo a matriz de pixels com os valores correspondentes. Finalmente, retorna a matriz de pixels.

O código também inclui um exemplo de uso, onde ele lê uma imagem chamada 'ufabc.pbm' e imprime os pixels da imagem.

````
# ler uma image no formato PBM
def leiaImagemPBM(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    # Verifica se o arquivo é uma imagem PBM válida
    if lines[0].strip() != 'P1':
        raise ValueError('O arquivo não é uma imagem PBM válida.')

    # Obtém as dimensões da imagem
    vet = lines[1].split()
    width, height = [int(vet[i]) for i in range(len(vet))]
    print(width, height)

    # Inicializa a matriz de pixels
    pixels = [[0]*width for _ in range(height)]

    # Percorre as linhas da imagem e preenche a matriz de pixels
    for i in range(2,height+1):
        row = lines[i].strip()
        for j in range(len(row)):
            pixels[i-2][j] = int(row[j])

    return pixels

# Exemplo de uso
filename = 'ufabc.pbm'
pixels = leiaImagemPBM(filename)

# Imprime a matriz de pixels
for i in range(len(pixels)):
  for j in range(len(pixels[i])):
    print(pixels[i][j],end='')
  print()
````

O método `dilata` a seguir recebe uma lista de pixels e aplica a operação de dilatação, que expande regiões com valor 1 na imagem. Para cada pixel $(i,j)$, o método encontra o valor máximo em uma vizinhança de 3x3 e substitui na imagem resultante `dil` de $(i,j)$ pelo valor máximo encontrado.







````
def dilata(pixels):
    height,width = len(pixels),len(pixels[0])
    dil = [[0]*width for _ in range(height)]

    for i in range(height):
        for j in range(width):
            max = pixels[i][j]
            for x in range(-1,2):
              for y in range(-1,2):
                viz_i = i + x
                viz_j = j + y
                if 0<=viz_i and viz_i < height and 0<=viz_j and viz_j < width:
                  if pixels[viz_i][viz_j] > max:
                    max = pixels[viz_i][viz_j]
            dil[i][j] = max

    return dil

# Exemplo de uso
dil = dilata(pixels)

# Imprime a matriz de pixels dilatada
for i in range(len(dil)):
    for j in range(len(dil[i])):
        print(dil[i][j], end='')
    print()
````
